# Base para estatística

## Amostra e população

### População

Termo usado para se referir a "todos os indivíduos"

Em um estudo estatístico, a população depende do sentido do estudo, por exemplo:

- se quero saber a opinião dos meus clientes sobre o atendimento durante a compra, então a população é "todos os clientes"

- se quero saber a opinião dos meus clientes sobre o SAC, então a população é "todos os clientes que passaram pelo SAC"

- se quero saber a o salário médio do brasileiro, então a população é todos os "brasileiros epregados"

Mas realizar uma análise com base na populção nem sempre e viável, pode ser impossível de coletar ou processar os dados.

 ### Amostra

Termo usado para se referir a "uma parte da população"

É uma quantidade relevante de indivíduos que quando analisados irão gerar inferências sobre a população

Mas adiante veremos sobre como desenvolver uma amostra de forma ideal, por enquanto imagine que a nossa amostra representa o todo, seria algo semelhante a ideia de *pesquisa de intenção de votos*.

## Amostragem

Termo usado para se referir ao processo de "extração da amostra a pratir da população"

Existem diferentes formas de gerar uma amostra para um estudo estatístico, a diferença está principalmente no "como fazer", mas também ímplica na representatividade que a amostra tem sobre a população.

 ### Técnicas de amostragem

As mais relevantes para ciência de dados.

##### Amostragem aleatória simples

É o processo mais comunmente ultilizado (de maneira não científica), consiste em retirar indivíduos ao acaso da população, isto é, de forma aleatória como um sorteio, onde todos tem a mesma chance de serem sorteados.

o problema com esse processo é que ele só é funcional quando a minha população em questão não é praticamente homogênia (os indivíduos não possuem caracteristicas que possam definir grupos ou categorias dentro da população).

Essa técnica pode não funcionar por exemplo para uma população de clientes, pois existem diferentes perfis de clientes, pode ser que determinado perfil não receba a devida relevância na amostra, sendo assim ela não representa todas as categorias naquela população, embora tenha sido feita de forma aleatória.
##### Amostragem estratificada

Esta técnica visa a representatvidade , segmentando a população em grupos ou extratos de acordo com as caracteristicas dos indivíduos, por exemplo podemos :

- segmentar uma população alunos de ensino médio de em extratos de masculino e feminino
- segmentar uma população atletas de judô de acordo com a idade ou peso de cada indivíduo criando extratos como "peso pena", "categoria mirim"

Uma vez que a populalção foi extratificada, aplicamos a mesma técnica da amostragem aleatória simples **dentro de cada extrato**, somando as amostras de cada extrato teremos a amostra da população.

Existem duas vertentes nessa técnica, as amostragens extratificadas **simples** e  **proporcionais**

A amostra simples retira o mesmo tanto de indivíduos de diferentes extratos, não se importando com o tamanho de cada extrato.

A amostragem proporcional analisa cada extrato em relação a população e retira indiviuos conforme a relevância do extrato a que eles pertencem.

## Estudos estatísticos

Temos dois principais tipos de estudos estatisticaos, **observacional** e **experimental**

O estudo **observacional** usa uma amostra como base de dados, e normalmente visa explorar caracteristicas dessa amostra num processo investigativo.

Já o estudo **experimental** cria um ambiente perfeito para realizar um teste, tendo o intuito de averiguar de maneira empírica os resultados.

## Variáveis estatísticas

A variável é uma informação a respeito do indivíduo que está presente no conjunto de dados e pode ser usada para analises, como por exmeplo:

- variável altura
- variável idade
- variável sexo
- variável classe social

Todos os individuos apresentam esses dados próprios, suas caracteristicas...

As variáveis são agrupadas em 2 grupos e cada grupo possui mais duas separações, mas a associuação é muito simples

#### Variáveis quantitativas

Estas variáveis são expressas por valores númericos, e podem ser de dois tipos

##### Discretas

As discretas normalmente surgem a partir de uma contagem, portanto só podem assumir valores inteiros.

ex: 12, 3, 1000

##### Contínuas

As contínuas normalmente surgem a partir de medições, por isso podem assumir valores quebrados, pois usam uma referencia para medir a determinada caracteristica do indivíduo.

ex: 12.0, 3.755, 0.564

#### Variáveis qualitativas
Essas variáveis são normalmente palavras, essas caracteristicas são agrupadas em categorias, como por exemplo o genero masculin ou feminino

##### Nominal

As categorias não possuem hierarquia, isto é não podem ser ordenadas.
ex.: sexo, raça, sobrenome

##### Ordinal

Possuem hierarquia e podem ser ordenadas.

ex.: faixa etária, classe social, nível de alguma coisa



> É importante saber que o tipo da variável pode mudar dependendo de como eu trato ela durante minha análise, por exemplo o "id da venda" pode ser algo como '738219' e eu tratar essa informação como se fosse o "nome da venda", então mesmo que tenha números nessa variável pode ser considerada qualitativa.

# Medidas de tendência central

## Média

A média pe a medida de tendência central mais popular, ela visa a ideia de representar o valor médio de um conjunto de dados, de forma que se esse conjunto possui valores grandes e pequenos esses valores se compensam...

A média é obtida através da soma de todos os valores de um conjunto de dados dividido pela quantidade de valores que esse conjunto possui, por exemplo:
 
6 alunos tiraram a seguinte nota em portugues:
5.5 - 3.5 - 7.0 - 9.0 - 8.5 - 2.5

A média desses 6 alunos foi de 6.0 (36/6)

A média serve para nos permitir analisar de modo superficial um dataset, pois ela não revela muita coisa.

> Estatística: a ciência que diz que se eu comi um frango e tu não comeste nenhum, teremos comido, em média, meio frango cada um. —  Dino Segre

## Mediana

A mediana busca representar  de foma visual e quantitativa um data.

Para obtermos a mediana de um conjunto de dados precisamos ordena-lo do menor para o maior, a mediana busca o valor que represente o "meio dessa fila"

A pode ser um dos números dessa fila, assim como podemos precisar realizar a média dentre os dois números do meio, veja os exemplos a seguir:


primeira situação:
1 - 2 - 3 - 4 - 5
A mediana = 3, pois ele divide o conjunto de dados ao meio


segunda situação:
1 - 2 - 3 - 4 - 5 - 6
A mediana = 3.5 (média de 3 - 4) , pois nenhum número ocupa exatamente o meio do conjunto de dados

> A melhor forma de entender a mediana e pensar que ea assume um valor, e que metade dos dados serão maiores que este valor e a outra metade serão menores.   1 - 2 - 3 (mediana) 4 - 5 - 6

## Moda

A mais simples de todas as medidas de tendência central, Moda é o número que mais se repete em um conjunto de dados.

Para obter a moda basta observar o número que mais se repetiu...

- se resultar em um empate então os dois valores podem ser a moda, 

- se nenhum número se repetir então não existe moda


exemplo:

1 - 1 - 2 - 2 - 3 - 2 - 1 - 3 - 1

o número 1 é a moda pois se repetiu mais.

# Medidas de dispersão

## Amplitude total

A amplitude total que nos passar a nossão do quanto o conjunto de dados pode variar

A amplitude total é a subtração do menor número pelo maior . é uma ferramenta de entendeer o quão dstante os dados podem estar uns dos outros.

exemplo:

1 - 2 - 4 - 6 - 8 - 9 - 12 - 2 - 4 - 7 - 7 - 9 - 14

maior = 14

menor = 1

a amplitude geral desse dataset = 13, essa é a maior variação que pode haver a partir do menor ou do maior, nenhum número se distância do outro mais que 13 "casas".

## Variância σ²

A variância mostra, em média, quão distante cada valor de um conjunto está da média aritimética.

> Não vou exibir o cálculo agora pois é irrelevante neste momento, neste nível de estudo o melhor a fazer é a abstração.

Saiba que a variância resulta num valor numérico ruim para fazer comparções e que pode ser menos usual que o desvio padrão estudado a diante, exemplo de valores de variância:

- [3, 12, 8, 15, 7] variância = 51.0

- [11, 16, 5, 18, 7] variância = 90.36


Mas o que esse valores querem dizer? esses valores deveriam ser mais palpáveis, como 7 ou 8...

## Desvio-padrão σ

E a média do quanto os termos se afastam da média aritimética daquele conjunto, quanto maior o desvio padrão mais dispersos estão os dados, quanto menor mais próximos eles estão.

Os calculos dos dois são parecidos parecido, porém o desvio padrão dá um passo a mais para converter a variância num valor palpável, ele apenas é a raiz quadrada da variância.

Resumindo, o desvio padrão e a variância são duas formas de medir a mesma coisa, a dispersão dos dados.

exemplos:

- [10, 20, 19, 16, 5] dp ~= 12

- [11, 13, 1, 10, 4] dp ~= 6

> Note que os resultados são aproximados pois estamos arredondando raizes "quebradas".

# Medidas separatrizes

## Quartis

Os quartis servem para separar um dataset depois de ordenado, a "quartificação" de um dataset divide-o em 4 partes, onde cada parte deve conter aproximadamente 25% dos itens daquele conjunto (1/4).



A forma como encontramos os pontos exatos que dividem o conjunto em quartis é basicamente usando as medianas.

AS, pois é preciso encontrar a mediana, que representa o Q2 (segundo quartil) e apartir disso encontrar a mediana da primeira metade do conjunto , Q1, e da segunda parte do conjunto , Q3.

Temos algo bem visual aqui, a ideia é extratificar o dataset.

## FIQ

Faixa interquartil, ela tenta medir a dispersão dos 50% ao centro da dispersão, a forma de fazer isso é calculando a amplitude total entre o primeiro quartil e o terceiro quartil, isto é **Q3 - Q1 = FIQ**.

A ideia é avaliar a parte central do conjunto de dados, parte esta que pode ser muito útil para analisees pois está afastada possíveis valores exagerados presentes nos extremos (outlier), que falaremos mais a diante.

## Outliers

A tradução de outlier pode ser algo como "um ponto fora da curva" ou seja, algo que se afasta do cumum.

Em estatística um conjunto que possui um dados afastado dos demais é considerado um outlier,exemplo:

Imagine que você está analisando a idade de alunos do ensino médio e tem uma amplitude total que no geral varia entre 14 e 18 anos , mas derrepente enxerga o valor 53, este é um outlier neste momento pois ele se afasta demais dos demais valores. se fizermos um calculo de média aritimética este valor pode **distorcer a nossa análise**.

E é justamente essa a preocupação com os outliers na estatística, deles distorcerem a análise. Outra ultiidade na identificação de outliers é que pode ser o caso de fazer uma investigação sopbre o indiviuo.


Então uma vez que se entende o que é um outlier, é preciso saber como definir exatamente o que é ou não um outlier para determiinado conjunto de dados (uma formula matemática)

Existe uma solução concenso na comunidade estatística, a regra de *1,5 . FIQ*

#### Regra 1,5 . FIQ

Esta regra diz que serão considerados outliers todos os valores :

**acima** de Q3 + (1,5 . FIQ) 

**abaixo** de Q1 - (1,5 . FIQ) 

criando assim limites para que esse conjunto de dados tenha ou não outliers

A distribuição normal ou Gaussiana é a maneira mais "natural" que os dados podem se dispor, isto porque quando observamos dados com muitos elementos estatisticos dados sobre fenomenos naturais, normalmente eles seguem este padrão... **"Muitos valores perto do centro e poucos valores distantes"**

 A distribuição ser normal ou não irá mudar a forma como realizamos alguns testes e quais dados estatísticos iremos usar de *input* para outros metodos estatícos.
